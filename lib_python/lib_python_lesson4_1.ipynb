## Задание 2

```python
t = np.linspace(0, 10, num=51)
f = np.cos(t)
```

```python
plt.plot(t,f, 'g')
# plt.plot(t,f, color = 'green')
plt.title("График f(t)")
plt.xlabel("Значения t")
plt.ylabel("Значения f")
plt.axis([0.5, 9.5, -2.5, 2.5])
plt.show
```

# Задание 3

```python
#Подготовка данных для графиков
x = np.linspace(-3, 3, 51)

y1 = x**2
y2 = 2*x + 0.5
y3 = -3*x - 1.5
y4 = np.sin(x)
# y1 = [el**2 for el in x]
# y3 = list(map(lambda el: (-3 * el - 1.5) , x))
```

```python
fig, ax = plt.subplots(nrows = 2, ncols = 2)
fig.subplots_adjust(wspace=0.3, hspace=0.3)
ax[0, 0].plot(x, y1)
ax[0, 1].plot(x, y2)
ax[1, 0].plot(x, y3)
ax[1, 1].plot(x, y4)

[ ax[ el[0],el[1] ].set_title(el[2]) for el in  [(0,0,"График y1"), (0,1,"График y2"), (1,0,"График y3"), (1,1,"График y4")]]

fig.set_size_inches(8, 6)
```

# Задание 4

```python
import matplotlib.pyplot as plt
import pandas as pd
plt.style.use('fivethirtyeight')
```

```python
data = pd.read_csv('../data/creditcard.csv')
```

```python
data
```

```python
classCount = data.value_counts('Class')
classCount
```

```python
classCount.plot(kind = "bar")
plt.show()
classCount.plot(kind = "bar", logy = True)
plt.show()
```

```python
v1_BadCounts = data[data['Class']==1].value_counts('V1')
v1_OkCounts = data[data['Class']==0].value_counts('V1')
v1_BadCounts
```

```python
plt.hist([v1_BadCounts,v1_OkCounts], log='True')
plt.legend(["v1_BadCounts","v1_OkCounts"])
plt.show()
```
