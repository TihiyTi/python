```python
import math
import numpy as np
import matplotlib.pyplot as plt

def f(k,x):
    return math.cos(k*x)
```

```python
points = 25
res = [f(1,math.pi/16*i) for i in range(points)]
res2 = [f(3,math.pi/16*i) for i in range(points)]
plt.plot(list(range(points)), res, res2)
```

```python
x = np.linspace(0, 10)
plt.plot(x, np.cos(1*x), np.cos(3*x))
```
